# CONVERTIR A VARIABLES DUMMIES

```python
datos = pd.get_dummies(Laptops['RAM'])
datos2 = pd.get_dummies(Laptops['tipo'])
datos = pd.concat([Laptops,datos, datos2], axis = 1)
print(datos.columns)
```

# PREPARAMOS EL MODELO PARA PREDECIR PRECIOS

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# DATOS DUMMIES DE RAM EN DATOS 
lm = LinearRegression()
X = datos[['pulgadas', '12GB', '16GB',
       '24GB', '2GB', '32GB', '4GB', '64GB', '6GB', '8GB',
       '2 in 1 Convertible', 'Gaming', 'Netbook', 'Notebook', 'Ultrabook',
       'Workstation']]

y = datos['precio_euros']
```

```python
#lm = LinearRegression()
#X = datos_dummies.drop('precio_euros', axis=1)  # Eliminar la columna 'precio_euros'
#y = datos_dummies['precio_euros']
```

```python
X
```

```python
y
```

# ENTRENAMOS EL MODELO 

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y, test_size = 0.2, random_state = 1) #DIVISION DE DATOS 
```

```python
lm.fit(X_train, y_train)
```

# PREDICCI√ìN 

```python
prediccion =  lm.predict(X_test)
```

# COMPARACI√ìN 

```python
resultado = {'Real':y_test, 'Prediccion':prediccion}
resultado = pd.DataFrame(resultado)
resultado.head(50)
```

# EVALUACI√ìN DEL MODELO 

```python
from sklearn.metrics import mean_absolute_error
```

```python
MAE = mean_absolute_error(y_test, prediccion)
MAE
```
